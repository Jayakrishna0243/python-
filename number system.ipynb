# 3.decimal number system

```python
d=10
print(d)
bin(d)
```

```python
v= 10
binary = bin(v)
print("Decimal:", v)
print("Binary:", binary)
```

```python
a= 345
binary = bin(a)
print("Decimal:", a)
print("Binary:", binary)
```

# 4.Hexdecimal number system

```python
hex(20)
```

```python
hex(5000)
```

```python
print(hex(0)) #0*0
print(hex(1)) #0*1
print(hex(2)) #0*2
print(hex(3)) #0*3
print(hex(4)) #0*4
print(hex(5)) #0*5
print(hex(6)) #0*6
print(hex(7)) #0*7
print(hex(8)) #0*8
print(hex(9)) #0*9
print(hex(10))#0*a
print(hex(11))#0*b
print(hex(12))#0*c
print(hex(13))#0*d
print(hex(14))#0*e
print(hex(15))#0*f
print(hex(16))#10 16/16=0==10
print(hex(17))#11  17/16=1==11
```
