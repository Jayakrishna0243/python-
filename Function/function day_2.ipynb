# 5. Variable length argument or kwargs

```python
#variable length argument
def sum(a,b):
    c=a+b
    print(c)
sum(5,6,7,8)
```

```python
def sum(a,*b):
    c=a+b
    print(c)
sum(5,6,7,8)
```

```python
def sum(a,*b):
    c=a+b
    print(c)
sum(5,6,7,8)
```

```python
def sum(a,*b):
    # c=a+b
    print(type(a))
    print(type(b))
    print(a)
    print(b)
sum(5,6,7,8)
```

```python
# now we can apply a FOR loop

def sum(a,*b):
    c=a
    for i in b:
        c=c+i
    print(c)
sum(5,6,7,8)
```

```python
def sum(a,*b):
    c=a
    for i in b:
        c=c+i
    print(c)
sum(5,6,7,8,9,10,-1,-2,100)
```

```python
def sum(a,*b):
    c=0
    for i in b:
        c=c+i
    print(c)
sum(5,6,7,8)
```
