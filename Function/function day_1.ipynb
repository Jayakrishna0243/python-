# 1.positional arguments

```python
def person(name ,age):
    print(name)
    print(age)
person('Nit')
```

```python
def person(name):
    print(name)
    print(age)
person('Nit',23)
```

```python
def person(name ,age):
    print(name)
    print(age)
person('Nit',23)
```

```python
def person(age,name):
    print(name)
    print(age)
person('Nit',23)
```

```python
def person(age,name):
    print(name)
    print(age+1)
person('Nit',23)
```

```python
def person(age,name):
    print(name)
    print(age+1)
person('Nit',23)
```

# 2. Keyword argument 

```python
def person(age,name):
    print(name)
    print(age+1)
person(name='Nit',age=23)
```

```python
def person(age,name,mob):
    print(name)
    print(age+1)
    print(mob)
person(name='Nit',age=23,12345)
```
